```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

```python
x = np.arange(0, 10, 1) # x nomli o'zgaruvchi yaratib oldim
sin = np.sin(x) # sin degan o'zgaruvchi yaratib oldim
cos = np.cos(x)

plt.figure(figsize= (12,6))
plt.subplot(2,1,1) # 2 qator va 1 ustunlik qilib 1 inchi grafikka chiz

# sin(x) grafigini chizamiz
plt.plot(x, sin,       # x va y kordinatalarni berdim
         linewidth = 2, # chiziq qalinligini berdim
         linestyle = '--', # chiziq turini berdim
         color = 'b', # chiziq rangini berdim
         alpha = 0.7, # chiziq shaffofligini berdim
         marker = 'o') # burilish nuqtalarini berdim
plt.title("sin(x) grafigini chizamiz") # grafikni nomladim
plt.grid(color = 'tomato', alpha=0.7, linestyle = '-') # koordinata setkalariga parametr berdim

#cos(x) grafigini chizamiz

plt.subplot(2,1,2)
plt.plot(x, cos,
         linewidth = 2,
         linestyle = '--',
         color = 'r',
         alpha = 0.7,
         marker = 'h')
plt.title("cos(x) grafigini chizamiz")
plt.grid(color = 'tomato', alpha=0.7, linestyle = '-')

plt.show()
```

## `seaborn` va `subplot`

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

#### Bu safar `subplots` funksiyasidan foydalanamiz.

`subplots` funksiyasi 2 ta obyekt qaytaradi:
1. Umumiy chizma obyekti
2. Chizma tarkibidagi grafiklar ro'yxati

Biz quyida, qulaylik uchun, bu ikkalasini `fig` (figure) va `ax` (axes) o'zgaruvchilariga saqlab olamiz:

```python
fig, ax = plt.subplots(2, 1, figsize=(10,10))
# Umumiy chizma nomini beramiz:
fig.suptitle("Grafik 1")
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, x=df['year'], y=df['price'], hue="fuelType", s=50)
plt.show()
```

2 o'lchamli grafik

```python
fig, ax = plt.subplots(2, 2, figsize=(15,10))
# Umumiy chizma nomini beramiz:
fig.suptitle("Grafik 2")
sns.histplot(ax=ax[0, 0], data=df, x='price')
sns.histplot(ax=ax[0, 1], data=df, x='mileage')
sns.scatterplot(ax=ax[1, 0], data=df, x=df['year'], y=df['price'], hue="fuelType", s=50)
sns.scatterplot(ax=ax[1, 1], data=df, x=df['year'], y=df['mileage'], hue="fuelType", s=50)
# Har bir grafik uchun nom:
ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("Price vs Year correlation")
ax[1,1].set_title("Mileage vs Year correlation")
plt.show()
```
